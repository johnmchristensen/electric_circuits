```python
import numpy as np

from ac_circuit_elements import from_phasor, to_phasor_string
from power import calc_max_average_power_resistive
from circuit_parallel import calc_par_imp

v = from_phasor(120, np.deg2rad(60))
r1 = 80
l = 60j
r2 = 90
c = -30j

z_thevenin = calc_par_imp(c, r2, l + r1)
print("z_thevenin = ", to_phasor_string(z_thevenin))

v_thevenin = v * calc_par_imp(r2, c) / (calc_par_imp(r2, c) + l + r1)
print("v_thevenin = ", to_phasor_string(v_thevenin))

r_load = abs(z_thevenin)
print("r_load = ", r_load)

i_thevenin = v_thevenin / (z_thevenin + r_load)
print("i_thevenin = ", to_phasor_string(i_thevenin))

p_max_average_absorbed = calc_max_average_power_resistive(i_thevenin, r_load)
print("Max Average Power: ", p_max_average_absorbed)
```
