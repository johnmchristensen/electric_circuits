```python
from ac_circuit_elements import to_phasor_string
import sympy

v = 165 # volts
r = 40 # ohms
l = 20j # ohms
c = -10j # ohms

i0, i1, i2, i3 = sympy.symbols('i0 i1 i2 i3')
eq = [sympy.Eq(i0 * r + i0 * l - i1 * l, 0),
      sympy.Eq(i1 * l + i1 * c - i0 * l - i2 * c, 0),
      sympy.Eq(i2 * c + 4 * i0 + i2 * c - i1 * c - i3 * c, 0),
      sympy.Eq(i3 * c + 1 - i2 * c, 0)]
sol = sympy.solve(eq, [i0, i1, i2, i3])
print(sol)

i_th = -sol[i3]
print("i (thevenin) = " , to_phasor_string(i_th), " amps")

z_th = (1 / i_th).evalf()
print("z (thevenin) = ", z_th, " ohms")

r_load = abs(z_th)
print("r (load) = ", r_load, " ohms")
```

```python
i4, i5, i6 = sympy.symbols('i4 i5 i6')
eq2 = [sympy.Eq(-v + r * i4 + l * i4 - l * i5, 0),
      sympy.Eq(l * i5 + c * i5 - l * i4 - c * i6, 0),
      sympy.Eq(4 * i4 + c * i6 + c * i6 - c * i5, 0)]
sol2 = sympy.solve(eq2, [i4, i5, i6])
print(sol2)

v_th = sol2[i6] * c
print("v (thevenin) = ", to_phasor_string(v_th), " volts")

p_max = 0.5 * abs(v_th / (z_th + r_load))**2 * r_load
print("p (max) = ", p_max, " watts")
```
