```python
p_a = 2000 # W
pf_a = 0.8 # lagging
theta_a = np.arccos(pf_a)
s_a = p_a / theta_a
print(s_a)
print(np.rad2deg(theta_a))
```

```python
pf_b = 0.4 # leading
theta_b = np.arccos(pf_b) * -1
print(np.rad2deg(theta_b))
```

```python
p_c = 1000 # W
q_c = 500 # VAR
pf_c = p_c / q_c
print(pf_c)
```

```python
s_b = 3 * np.exp(1j * theta_b)
s_c = 1000 + 500j
s = s_a + s_b + s_c
print(s)
theta = np.angle(s)
pf = np.cos(theta)
print(pf)
```

```python
v = 120 * np.exp(1j * np.deg2rad(45))
i_conj = s / v
print(i_conj)
i = i_conj.conjugate()
print(abs(i), "@", np.rad2deg(np.angle(i)))
```
