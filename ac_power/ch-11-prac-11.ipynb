```python
import numpy as np

from ac_circuit_elements import from_phasor, to_phasor_string

V_rms = from_phasor(110, np.deg2rad(85))
I_rms = from_phasor(400 * .001, np.deg2rad(15))


S = V_rms * I_rms.conjugate()
print("Complex Power: ", to_phasor_string(S), " VA")
s = abs(S)
print("Apparent Power: ", s, "VA")

p = S.real
q = S.imag

print("Real Power: ", p, "W")
print("Reactive Power: ", q, "VAR")

pf = np.cos(np.angle(V_rms) - np.angle(I_rms))
print("Power Factor: ", pf)

Z = V_rms / I_rms
print("Load Impedence: ", Z, " Ohms")
```
