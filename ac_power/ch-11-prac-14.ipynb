```python
from ac_circuit_elements import to_phasor_string
from power import calc_s_from_power_and_power_factor, calc_power_factor_from_s

p_1 = 2 * 10**3 # 2 kW
pf_1 = 0.75 # leading
s_1 = calc_s_from_power_and_power_factor(p_1, pf_1, is_leading=True)
print("s_1: ", to_phasor_string(s_1))

p_2 = 4 * 10**3 # 4 kW
pf_2 = 0.95 # lagging
s_2 = calc_s_from_power_and_power_factor(p_2, pf_2, is_leading=False)
print("s_2: ", to_phasor_string(s_2))

s = s_1 + s_2
print("S: ", s)
print("Power Factor: ", calc_power_factor_from_s(s))
```
