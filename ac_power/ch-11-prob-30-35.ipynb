```python
import sympy as sp

from power import piecewise_rms
from sympy.plotting import plot

display("Exercise 11.30")
period = 4
t = sp.Symbol('t')
f = [2 * t, -1]
ranges = [(0, 2), (2, 4)]

v_rms = piecewise_rms(f, ranges, period, t)
print("v (rms) = ", v_rms, " Volts")
```

```python
display("(b)")
f = [4 + 3 * sp.cos(5 * t)]
period = 2 * sp.pi / 5
ranges = [(0, period)]
v_rms = piecewise_rms(f, ranges, period, t)
print("v (rms) = ", v_rms.evalf(), " Volts")
```

```python
display("(c)")
f = [8 - 6 * sp.sin(2 * t)]
period = sp.pi
ranges = [(0, period)]
i_rms = piecewise_rms(f, ranges, period, t)
print("i (rms) = ", i_rms.evalf(), " Amps")
```
