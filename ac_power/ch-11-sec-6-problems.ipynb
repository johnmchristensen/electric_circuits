```python
import numpy as np

from power import piecewise_rms

display("Problem 11.44")
v = 110 # v (rms)
f = 60 # hz
s = 120 # VA
pf = 0.707
leading = False # v > i

p = s * pf
s_angle_rad = np.arccos(pf)
q = s * np.sin(s_angle_rad)
s_complex = p + q * 1j
print("S = ", s_complex)
```

```python
from ac_circuit_elements import from_phasor, to_phasor_string, to_angular_frequency, to_capacitor_impedance, \
    to_inductor_impedance

i = s_complex / v
print("i = ", to_phasor_string(i), " amps (rms)")
```

```python
z = v / i
print("z = ", z, " ohms")
```

```python
w = to_angular_frequency(f)
l = abs(z.imag / w)
print("l = ", l * 1000, " mh")
```

# Problem 11.44

```python
w = 2000
c = 40 * 10**-6 # micro-farads
l = 60 * 10**-3 # mh
z_c = to_capacitor_impedance(w, c)
z_l = to_inductor_impedance(w, l)

print("z (inductor) = ", z_l, " ohms")
print("z (capacitor) = ", z_c, " ohms")
```

```python
import sympy as sp

v_s = 100 # volts
r_30 = 30
r_20 = 20

v_n, i_x = sp.symbols ('v_n i_x')
eq = [sp.Eq((v_n - v_s) / (r_30 + z_c) + v_n / z_l + (v_n - 4 * i_x) / r_20, 0),
      sp.Eq(i_x, v_n / z_l)]
sol = sp.solve(eq, (v_n, i_x))
print(sol)
```

```python
i = (sol[v_n] - v_s) / (r_30 + z_c)
print("i = ", to_phasor_string(i), " amps (rms)")

s = (v_s * i).evalf()
print("complex power = ", s, " VA")
```

# Problem 11.45

```python
t = sp.Symbol('t')
v = 20 + 60 * sp.cos(100*t) # volts
i = 1 - 0.5 * sp.sin(100*t)

v_rms = piecewise_rms([v], [(0, 2*sp.pi)], 2 * sp.pi, t).evalf()
i_rms = piecewise_rms([i], [(0, 2 * sp.pi)], 2 * sp.pi, t).evalf()

print("v (rms) = ", v_rms, " volts")
print("i (rms) = ", i_rms, " amps")
```

```python
p = v * i
p_avg = sp.integrate(p, (t, 0, 2 * sp.pi)) / (2 * sp.pi)
print("p (average) = ", p_avg, " watts")
```

# Problem 11.46

```python
from power import calc_power_factor


def solve_problem_11_46(part, v, i):
    s = v * i.conjugate()
    pf, leading = calc_power_factor(v, i)
    p = s.real
    q = s.imag
    s_apparent = abs(s)

    print(part)
    print("s = ", s, " VA")
    print("s (apparent) = ", s_apparent, " VA")
    print("p = ", p, " Watts")
    print("q = ", q, " VAR")
    print("pf  ", pf, " ", "leading" if leading else "lagging")
    print("")

v = from_phasor(220, np.deg2rad(30)) # volts (rms)
i = from_phasor(0.5, np.deg2rad(60)) # amps (rms)

solve_problem_11_46('a', from_phasor(220, np.deg2rad(30)), from_phasor(0.5, np.deg2rad(60)))
solve_problem_11_46('b', from_phasor(250, np.deg2rad(-10)), from_phasor(6.2, np.deg2rad(-25)))
solve_problem_11_46('c', 120, from_phasor(2.5, np.deg2rad(-15)))
solve_problem_11_46('d', from_phasor(160, np.deg2rad(45)), from_phasor(8.5, np.deg2rad(90)))
```

# Problem 11.47

```python
def solve_problem_11_46_with_v_and_i(part, v, i):
    v_rms = v / np.sqrt(2)
    i_rms = i / np.sqrt(2)
    s = v_rms * i_rms.conjugate()
    print(part)
    print("|S| = ", abs(s), " VA")
    print("P = ", s.real, " W")
    print("Q = ", s.imag, " VAR")
    print("")

solve_problem_11_46_with_v_and_i('a', from_phasor(112, np.deg2rad(10)), from_phasor(4, np.deg2rad(50)))
solve_problem_11_46_with_v_and_i('b', 160, from_phasor(4, np.deg2rad(45)))
```
