```python
import numpy as np

from ac_circuit_elements import from_phasor, to_phasor_string, get_magnitude
from circuit_parallel import calc_par_imp
from power import calc_max_average_power_resistive

v = from_phasor(150, np.deg2rad(30))
r = 40
c = -30j
l = 20j

z_thevenin = calc_par_imp(r + c, l)
print("z_thevenin = ", z_thevenin)

v_thevenin = v * l / (l + r + c)
print("v_thevenin = ", to_phasor_string(v_thevenin))

r_load = get_magnitude(z_thevenin)
print("r_load = ", r_load)

i = v_thevenin / (z_thevenin + r_load)
print("I = ", to_phasor_string(i))

p_max = calc_max_average_power_resistive(i, r_load)
print("p_max = ", p_max)
```
