```python
import numpy as np

from ac_circuit_elements import from_phasor, to_phasor_string

q = 100 * 10**3 #kVAR
Z = from_phasor(250, np.deg2rad(-75))

theta_v = 0 # by definition
theta_i = np.deg2rad(75)
pf = np.cos(theta_v - theta_i)
print("Power Factor: ", pf)

S = from_phasor(q / np.sin(theta_v - theta_i), theta_v - theta_i)
print("Complex Power: ", to_phasor_string(S))

s = abs(S)
print("Apparent Power: ", s, " VA")

V_rms = np.sqrt(S * Z)
v_rms = abs(V_rms)
print("V_rms: ", v_rms, " V")
```
