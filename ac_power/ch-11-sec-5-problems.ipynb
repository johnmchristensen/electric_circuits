```python
import numpy as np
from power import calc_average_power, calc_power_factor, complex2rms, calc_reactive_power
from ac_circuit_elements import to_phasor_string
from circuit_parallel import calc_par_imp

display("Problem 11.38")
v = 220 * np.sqrt(2) # v
z = calc_par_imp(124, 20 - 25j, 90 + 80j)
print("z = ", z, " ohms")

i = v / z
print("i = ", to_phasor_string(i), " amps")

p_avg = calc_average_power(v, i)
print("(a) p_avg = ", p_avg, " watts")

q = calc_reactive_power(v, i)
print("(b) q = ", q, " VAR")

power_factor = calc_power_factor(complex2rms(v), complex2rms(i))
print("(c) power factor = ", power_factor)
```

```python
from power import calc_unity_capacitance, rms2mag, mag2rms
from ac_circuit_elements import to_angular_frequency

display("Problem 11.39")

w = to_angular_frequency(60)
v = rms2mag(220)
z_l = 4.2 + 3.6j # ohms
i = v / z_l

print("(a)")
p = calc_average_power(v, i)
pf = calc_power_factor(v, i)
q = calc_reactive_power(v, i)

print("p = ", p, " watts")
print("pf = ", pf)
print("q = ", q, " VAR")

c = calc_unity_capacitance(mag2rms(v), w, q)
print("c = ", c * 10**6, " micro-farad")
```
