```python
import sympy

from ac_circuit_elements import to_capacitor_impedance, to_inductor_impedance

w = 4 # rad/s
v = 10 # volts
r2 = 2 # ohms
r4 = 4 # ohms
c = 1/20 # farad
zc = to_capacitor_impedance(w, c) # ohms
l = 1 # henry
zl = to_inductor_impedance(w, l) # ohms

# part 1`- determine z_th by adding a 1-V excitation source.
v0, i1 = sympy.symbols('v0 i1')
eq = [sympy.Eq(v0 / r2 + v0 / zc + 0.5 * v0, 0),
      sympy.Eq((1 - v0) / r4 + 1 / zl - 0.5 * v0, i1)]
sola = sympy.solve(eq, [v0, i1])

z_th = 1 / complex(sola[i1])
print("Z (thevenin) = ", z_th, " ohms")

z_l = z_th.conjugate()
print("Z (load) = ", z_l, " ohms")
```
