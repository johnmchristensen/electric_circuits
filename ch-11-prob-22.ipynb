```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

t = sp.Symbol('t')
f = sp.Piecewise((4 * sp.sin(t), (t < sp.pi)),
                 (4 * sp.sin(t - sp.pi), (sp.pi < t)))
f_np = sp.lambdify(t, f, "numpy")


xt = np.linspace(0, 2 * np.pi, 1000)
yt = f_np(xt)

plt.plot(xt, yt)
```
