```python
import sympy

from power import calc_max_average_power

v = 12 # volts
r = 1 # ohms
l = 1j # ohms
c = -1j # ohms

v0, i_th = sympy.symbols('v_th i_th')
eq = [sympy.Eq(v0 / r + v0 / l + (v0 - 1) / c, 0),
      sympy.Eq(-2 * v0 + (1 - v0) / c - i_th, 0)]
sol = sympy.solve(eq, [v0, i_th])

z_th = (1 / sol[i_th]).evalf()
print("Z (thevenin) =", z_th, " Ohms")
z_l = z_th.conjugate().evalf()
print("Z (load) = ", z_l, " Ohms")
```

```python
v0a, v_th = sympy.symbols('v0a, v_th')
eq_b = [sympy.Eq((v0a - v) / r + v0a / l + (v0a - v_th) / c, 0),
        sympy.Eq((v_th - v0a) / c - 2 * v0a, 0)]
sol = sympy.solve(eq_b, [v0a, v_th])
print(sol)

p_max = calc_max_average_power(complex(sol[v_th]), complex(z_th))
print("Max average power: ", p_max, " Watts")
```
