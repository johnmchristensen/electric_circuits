```python
from calcParImp import calcParImp
from ac_circuit_elements import from_phasor, to_phasor, to_string
from IPython.display import Markdown, display
import numpy as npa
import sympy as sp

display(Markdown("# Circuit 1"))

i1a, i2a = sp.symbols('i1a i2a')
eqa = [sp.Eq(0, from_phasor(50, np.deg2rad(30)) + i2a * 4j + i2a * 6 - i1a * 4j),
       sp.Eq(0, i1a * 4j + i1a * 8 - i1a * 2j - i2a * 4j)]
sol = sp.solve(eqa, [i1a, i2a])
print("i1 = ", sol[i1a].evalf(), " = ", to_string(*to_phasor(sol[i1a])))
i0a = -sol[i1a].evalf()
print(f"i0' = {i0a} = {to_string(*to_phasor(i0a))}")
```

```python
display(Markdown("# Circuit 2"))

i = 10
v = sp.symbols('v')
eq = sp.Eq(i + v/6 + v/4j + v/(8-2j), 0)
sol = sp.solve(eq, v)
print("v: ", sol[0])

i0b = (sol[0] / (8-2j)).evalf()
print("i0b: ", i0b, "(" + to_string(*to_phasor(i0b)) + ")")
```
