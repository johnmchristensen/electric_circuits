```python
from circuit_parallel import calc_par_imp
from power import mag2rms
from ac_circuit_elements import from_phasor, to_inductor_impedance, to_capacitor_impedance, get_magnitude

v = from_phasor(8, -40) # volts
w = 2 # radians / sec
r1 = 1 # ohm
r2 = 2 # ohms
l = 3 # henry
zl = to_inductor_impedance(w, l)
c = 0.25 # farad
zc = to_capacitor_impedance(w, c)

z_parallel = calc_par_imp(zl, r2 + zc)
z = z_parallel + r1
v_r1 = v * r1 / z
v_r1_rms = mag2rms(get_magnitude(v_r1))

p_r1 = v_r1_rms ** 2 / r1
print("p_r1 = ", p_r1, " W")

v_parallel = v * z_parallel / z
v_r2 = v_parallel * r2 / (zc + r2)
v_r2_rms = mag2rms(get_magnitude(v_r2))
p_r2 = v_r2_rms ** 2 / r2
print("p_r2 = ", p_r2, " W")
```
