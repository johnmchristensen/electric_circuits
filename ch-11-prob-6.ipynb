```python
from power import complex2rms
from ac_circuit_elements import to_capacitor_impedance, to_inductor_impedance
import sympy

w = 10**3 # rad/s
i_s = 6 # A
r_50 = 50 # ohms
c = 40 * 10**-6 # micro F
zc = to_capacitor_impedance(w, c)
l = 20 * 10**-3 # milli-H
zl = to_inductor_impedance(w, l)
r_10 = 10 # ohms

v1, ix = sympy.symbols('v1 ix')
eq = [sympy.Eq(i_s, v1 / (r_50 + zc) + (v1 - 20 * ix) / (r_10 + zl)),
      sympy.Eq(ix, v1 / (r_50 + zc))]
sol = sympy.solve(eq, [v1, ix])

i_x_rms = complex2rms(sol[ix])
print(f"i_x_rms = {i_x_rms} A (RMS)")
p_50 = i_x_rms**2 * r_50
print(f"p_50 = {p_50} W ")
```
