```python
from ac_circuit_elements import to_phasor_string, get_magnitude
from circuit_parallel import calc_par_imp
from power import calc_power_factor

v_rms = 165

r1 = 10
r2 = 8
l = 4j
c = -6j

z = calc_par_imp(r2 + c, l) + r1
print("Impedance: ", to_phasor_string(z))

i_rms = v_rms / z
print("Current (rms): ", to_phasor_string(i_rms))

s = v_rms * get_magnitude(i_rms)
print("Apparent power: ", to_phasor_string(s))

pf = calc_power_factor(v_rms, i_rms)
print("Power factor: ", pf)
```
