```python
from ac_circuit_elements import from_phasor, to_phasor_string
import numpy as np

from circuit_parallel import calc_par_imp

v = from_phasor(120, np.deg2rad(50)) # volts
r100 = 100 # ohms
r40 = 40 # ohms
l = 30j # ohms
c = -10j # ohms
r50 = 50 # ohms
i = from_phasor(2, np.deg2rad(90)) # amps

z_th = calc_par_imp(r50, calc_par_imp(r100, r40 + l) + c)
print(z_th)
```

```python
import sympy

v1, v2 = sympy.symbols('v1 v2')
eq = [sympy.Eq((v1 - v) / r100 + v1 / (r40 + l) + (v1 - v2) / c, 0),
      sympy.Eq((v2 - v1) / c + v2 / r50, i)]
sol = sympy.solve(eq, [v1, v2])
v_th = sol[v2]
print("v (thevenin) = ", to_phasor_string(v_th), " V")
```

```python
r_load = abs(z_th)
print("r (load) = ", r_load, " ohms")
```

```python
i_th = v_th / (z_th + r_load)
p_max = 0.5 * abs(i_th)**2 * r_load
print("p (max) = ", p_max, " W")
```
