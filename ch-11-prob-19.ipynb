```python
import sympy

from ac_circuit_elements import to_phasor_string
from circuit_parallel import calc_par_imp

l = 1j # ohms
r3 = 3 # ohms
i = 33 # amps
r6 = 6 # ohms
c = -2j # ohms

zth = calc_par_imp(l + r3, r6) + c
print("zth = ", zth, " ohms")

vth = sympy.symbols('vth')
eq = sympy.Eq(vth / (l + r3) - i + vth / r6, 0)
sol = sympy.solve(eq, [vth])
print("vth = ", to_phasor_string(sol[0].evalf()), " volts")

rl = abs(zth)
print("r (load) = ", rl, " ohms")

i_l = sol[0] / (zth + rl)
print("i (load) = ", to_phasor_string(i_l.evalf()), " amps")

pmax = 0.5 * abs(i_l)**2 * rl
print("pmax (load) = ", pmax, " watts")
```
