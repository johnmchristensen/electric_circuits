```python
from power import piecewise_rms, calc_power_from_current_rms
import sympy as sp

period = 30
t = sp.Symbol('t')
f = [6 * t, -6 * (t - 10), 6 * (t - 20), -6 * (t - 30)]
ranges = [(0, 5), (5, 15), (15, 25), (25, 30)]

i_rms = piecewise_rms(f, ranges, period, t)
print("i (rms) = ", i_rms, " Amps")

r = 12 # ohms
p_avg = calc_power_from_current_rms(i_rms, r)
print("p (avg) = ", p_avg, " Watts")
```
