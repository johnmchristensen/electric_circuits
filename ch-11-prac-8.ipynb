```python
from matplotlib import pyplot as plt
import numpy as np
import sympy as sp

T = np.pi * 3
t = sp.Symbol('t')
f = 100 * sp.sin(t % sp.pi)
f_np = sp.lambdify(t, f, "numpy")
ts = np.linspace(0, T, 1000)
ys = f_np(ts)

plt.figure(figsize=(7,4))
plt.plot(ts, ys, label="i(t)")
plt.axvline(1, color='gray', linestyle='--', linewidth=1)
plt.xlim(0, T)
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlabel("t")
plt.ylabel("v(t)")
plt.title("full-wave rectified sine voltage")
plt.show()
```

```python
from power import calc_power_from_voltage_rms

v_rms = (sp.sqrt((1/T) * sp.integrate(f**2, (t, 0, T)))).evalf()
print("V_rms = ", v_rms)

p_rms = calc_power_from_voltage_rms(v_rms, 6)
print("p_rms = ", p_rms)
```
