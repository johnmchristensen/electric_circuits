```python
%env NGSPICE_LIBRARY_PATH=/opt/homebrew/Cellar/libngspice/45.2/lib/libngspice.dylib
%env KICAD9_SYMBOL_DIR=/Applications/KiCad/KiCad.app/Contents/SharedSupport/symbols/
%env KICAD_SYMBOL_DIR=/Applications/KiCad/KiCad.app/Contents/SharedSupport/symbols/
%env KICAD9_FOOTPRINT_DIR=/Applications/KiCad/KiCad.app/Contents/SharedSupport/footprints/

import numpy as np

def parallel(*resistances):
    resistances_np = np.array(resistances)
    inverted = 1 / resistances_np
    return 1 / inverted.sum()

def from_phasor(magnitude, angle_degrees):
    angle_radians = np.radians(angle_degrees)
    real = magnitude * np.cos(angle_radians)
    imag = magnitude * np.sin(angle_radians)
    return complex(real, imag)

def to_phasor(complex) -> (float, float):
    return abs(complex), np.degrees(np.angle(complex))
```

```python
rc = -2j
rl = 6j
r8 = 8

y_resistor_product = (rc * rl) + (rc * r8) + (r8 * rl)
ra = y_resistor_product / rc
rb = y_resistor_product / rl
rc = y_resistor_product / r8

print(ra, rb, rc)
```

```python
rd = parallel(4, rc)
re = parallel(10, ra)

print("RD = ", rd, " Ohms")
print("RE = ", re, " Ohms")
```

```python
rf = rd + re
print("RF = ", rf, " Ohms")
```

```python
rg = parallel(rf, rb)
rg
```

```python
z = 2 + rg
v = from_phasor(60, -30)
i = v / z
i_phasor = to_phasor(i)
i_phasor
```
