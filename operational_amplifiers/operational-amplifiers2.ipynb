```python
# Practice 5.3
r1 = 4 * 10**3 # ohms
rf = 280 * 10**3 # ohms
v1 = 45 * 10**-3 # volts
vo = -(rf / r1)*v1
print(vo)
```

```python
io = (v1 - vo) / rf
print(io)
```

```python
i_s, vo, r1, r2, r3 = sp.symbols('i_s vo r1 r2 r3')
v2 = sp.symbols('v2')
n1 = sp.Eq(i_s, (-v2) / r1)
n2 = sp.Eq((v2) / r1 + (v2 - vo) / r3 + v2 / r2, 0)
sol = sp.solve([n1, n2], [v2, vo])
print(sol)
```

```python
gain = sp.simplify(sol[vo] / i_s)

gain_proof = -r1 * (1 + r3 / r1 + r3 / r2)
print(sp.simplify(gain - gain_proof) == 0)
```

```python
# Practice 5.5
vs = 3 # volts
r4 = 4000 # ohms
r8 = 8000 # ohms
r5 = 5000 # ohms
r2 = 2000 # ohms
vi = vs * (r8 / (r4 + r8))
vo = sp.Symbol('vo')
n = sp.Eq((vi - vo) / r5 + vi / r2, 0)
sol = sp.solve([n], [vo])
print(sol)
```
