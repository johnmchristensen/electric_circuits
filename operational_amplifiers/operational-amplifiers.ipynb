```python
import numpy as np
import sympy as sp
a = 2 * 10**5
ri = 2 * 10**6 # ohms
ro = 50 # ohms
r5 = 5 * 10**3 # ohms
r40 = 40 * 10**3 # ohms
r2 = 20 * 10**3 # ohms
vs = 1 # volts
v1, vo, avd = sp.symbols('v1 vo avd')
n1 = sp.Eq(v1 / r5 + (v1 - vs) / r2 + (v1 - vo) / r40, 0)
n2 = sp.Eq((vo - avd) / ro + vo / r2 + (vo - v1) / r40, 0)
n3 = sp.Eq(avd, a * (vs - v1))
sol = sp.solve([n1, n2, n3], [v1, vo, avd])

vo_sol = sol[vo].evalf()
print(vo_sol)

g = sol[vo] / vs
print(g.evalf())

io = (sol[vo] - sol[avd]) / ro
print(io.evalf())
```

```python
r10 = 10 * 10**3 # ohms
r20 = 20 * 10**3 # ohms
vs, vo = sp.symbols('vs vo')
n = sp.Eq((0 - vs) / r10 + (0 - vo) / r20, 0)
sol = sp.solve(n, [vs, vo])
g = sol[0][1] / sol[0][0]
print(g)
```

```python
vs = 2 # volts
i = sp.Symbol('i')
n = sp.Eq((0 - vs) / r10 + i, 0)
sol = sp.solve(n, [i])
print(sol)
```
