```python
from power import calc_max_average_power
from ac_circuit_elements import from_phasor, to_phasor_string
from circuit_parallel import calc_par_imp
import numpy as np
import sympy

r30 = 30 # ohms
c = -10j # ohms
r40 = 40 # ohms
l = 20j # ohms
i_s = from_phasor(5, np.deg2rad(90)) # amps

z_th = calc_par_imp(r30 + c, r40 + l)
z_l = z_th.conjugate()
print("z_l = ", z_l, " Ohms")

vn, vl = sympy.symbols('vn vl')
eq = [sympy.Eq(-i_s + (vn - vl) / (c + r40) + (vn - vl) / (l + r30), 0),
      sympy.Eq(vl, 0)]
sol = sympy.solve(eq, [vn, vl])
print(sol)

v_left = (sol[vn]) * c / (c + r40)
v_right = sol[vn] * r30 / (r30 + l)
v_th = v_left - v_right
p_maxavg = calc_max_average_power(v_th, z_th)
print("P (max avg) = ", p_maxavg, " W")
```
