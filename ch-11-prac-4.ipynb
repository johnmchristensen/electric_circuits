```python
import numpy as np
import sympy as sp
from ac_circuit_elements import from_phasor, to_phasor_string, get_magnitude, get_angle_radians
from ac_power import calc_average_power
from ohms_law import calc_voltage
from sympy_extensions import solve_with_evalf

v1 = 40
r = 8
c = -2j
i = 4j
v2 = from_phasor(20, np.deg2rad(90))

i1, i2 = sp.symbols('i1 i2')
eq = [sp.Eq(-v1 + r*i1 + c*i1 - c*i2, 0),
      sp.Eq(v2 + c*i2 + i*i2 - c*i1, 0)]
sol = solve_with_evalf(eq, [i1, i2])
print(type(sol))

for i, v in sol.items():
    print(i, to_phasor_string(v))

print("P - V1: ", -calc_average_power(v1, sol[i1]))
print("p - R", calc_average_power(calc_voltage(sol[i1], r), sol[i1]))
print("P - V2: ", calc_average_power(v2, sol[i2]))
```
