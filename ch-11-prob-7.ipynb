```python
from power import complex2rms
from ac_circuit_elements import from_phasor
import numpy as np
import sympy

from circuit_parallel import calc_par_imp

vs = from_phasor(8, np.deg2rad(20))
r4 = 4
c = -5j
l = 5j
r10 = 10
z_eq = calc_par_imp(l, c + r10)

i0, v, v0 = sympy.symbols('i0 v v0')
eq = [sympy.Eq(i0, (vs - 0.1 * v0) / r4),
      sympy.Eq(v, 8 * i0 * z_eq),
      sympy.Eq(v0, v * r10 / (r10 + c))]
sol = sympy.solve(eq, [i0, v, v0])

v0_rms = complex2rms(sol[v0])
p = v0_rms**2 / r10
print(f"p = {p} W")
```
