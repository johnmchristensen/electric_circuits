```python
import numpy as np

from ac_circuit_elements import to_phasor_string, get_magnitude
from circuit_parallel import calc_par_imp

r1 = 20 # ohms
r2 = 30 # ohms
c = -10j # ohms
l = 20j # ohms
r3 = 60 # ohms
s_r3 = 240 # watts

v_rms_r3 = np.sqrt(s_r3 * r3)
print("v_rms @ r3: ", v_rms_r3, " V (rms)")

V_rms_b2 = v_rms_r3 * (r3 + l) / r3
print("v_rms @ b2: ", to_phasor_string(v_rms_b2), " V (rms)")

Zb = calc_par_imp(r2 + c, l + r3)
print("Zb: ", to_phasor_string(Zb), " Ohms")

V_rms = V_rms_b2 * (Zb + r1) / Zb
print("V_rms: ", to_phasor_string(V_rms), " V (rms)")

V_rms_r1 = V_rms * r1 / (Zb + r1)
print("V_rms_r1: ", to_phasor_string(V_rms_r1), " V (rms)")
S_r1 = abs(V_rms_r1) ** 2 / r1.conjugate()
print("S_r1: ", S_r1, " VA")

S_b1 = abs(V_rms_b2) ** 2 / (r2 + c).conjugate()
print("S_b1: ", S_b1, " VA")

S_b2 = abs(V_rms_b2) ** 2 / (l + r3).conjugate()
print("S_b2: ", S_b2, " VA")
```
