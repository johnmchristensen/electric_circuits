```python
from power import piecewise_rms, calc_power_from_voltage_rms
import sympy as sp

period = 12
t = sp.Symbol('t')
v = [4 * t, 4 * (t - 5), 4 * (t - 10)]
ranges = [(0, 2), (5, 7), (10, 12)]
v_rms = piecewise_rms(v, ranges, period, t)
print("v (rms) = ", v_rms)
```

```python
r = 2 # ohms
p_avg = calc_power_from_voltage_rms(v_rms, r)
print("p (avg) = ", p_avg, " Watts")
```
