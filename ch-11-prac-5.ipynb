```python
from ac_circuit_elements import to_phasor_string
from ac_power import calc_max_average_power
from circuit_parallel import calc_par_imp
from sympy_extensions import solve_with_evalf
import sympy as sp

r1 = 8
c = -4j
l = 10j
r2 = 5
ic = 12

z_thevenin = calc_par_imp(r2, r1 + c + l)
print("z_thevenin = ", to_phasor_string(z_thevenin))

v1, v2 = sp.symbols('v1 v2')
eq = [sp.Eq((v1 / (r1 + c)) + ((v1 - v2) / l), ic),
      sp.Eq(((v2 - v1) / l) + (v2 / r2), 0)]
sol = solve_with_evalf(eq, [v1, v2])

for i, v in sol.items():
    print(i, to_phasor_string(v))

z_load = z_thevenin.conjugate()
print("Zl for maximum average power: ", z_load)

p_max_average = calc_max_average_power(sol[v2], z_thevenin)
print("Max Avg Power: ", p_max_average)
```
