```python
import sympy as sp

from ac_circuit_elements import to_phasor

i0, i1, i2 = sp.symbols('i0 i1 i2')
eq = [sp.Eq(-60 + (10 - 4j + 5) * i0 + 4j*i1 + -5*i2, 0),
      sp.Eq((-4j + 8j) * i1 + (5 - 6j) * i2 - (-4j + 5) * i0, 0),
      sp.Eq(i2, i1 + 2.4)]

sol = sp.solve(eq, (i0, i1, i2))
print(sol[i0])
print(type(complex(sol[i0].evalf())))
print(to_phasor(sol[i0]))
```
