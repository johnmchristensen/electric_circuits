```python
from ac_circuit_elements import get_magnitude
from power import complex2rms
import sympy

i_s = 6 # A
l = 10j # ohms
c = -20j # ohms
r = 40 # ohms

v1, v2, i0 = sympy.symbols('v1 v2 i0')
eq = [sympy.Eq(i_s, v1 / l + (v1 - v2) / c),
      sympy.Eq(0.5 * i0, (v2 - v1) / c + v2 / r),
      sympy.Eq(i0, (v1 - v2) / c)]
sol = sympy.solve(eq, [v1, v2, i0])
print(sol)

p = get_magnitude(sol[v2])**2 / r
print(f"p = {p} W")
```
