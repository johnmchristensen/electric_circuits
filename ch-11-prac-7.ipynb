```python
from matplotlib import pyplot as plt
import numpy as np
import sympy as sp

T = 2

x, y = sp.symbols('x y')
f = sp.Piecewise((16 * x, (0 < x) & (x < 1)),
                 (-16 * x + 32, (1 < x) & (x < 2)),
                 (sp.nan, True))

f_np = sp.lambdify(x, f, "numpy")
xs = np.linspace(0, T, 1000)
ys = f_np(xs)

plt.figure(figsize=(7,4))
plt.plot(xs, ys, label="i(t)")
plt.axvline(1, color='gray', linestyle='--', linewidth=1)
plt.xlim(0, T)
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlabel("t")
plt.ylabel("i(t)")
plt.title("piecewise current")
plt.show()
```

```python
from power import calc_power

i_rms = (sp.sqrt((1/T) * sp.integrate(f**2, (x, 0, T)))).evalf()
print("I_rms = ", i_rms)

p_rms = calc_power(i_rms, 9)
print("p_rms = ", p_rms)
```
