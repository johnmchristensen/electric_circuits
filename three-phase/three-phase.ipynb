# Practice 2

```python
import sys

from pathlib import Path

sys.path.append(str(Path.cwd().parent))

from ac_circuit_elements import from_phasor, from_phasor_deg, to_phasor_string

van = from_phasor_deg(120, 30)
vbn = from_phasor_deg(120, -90)
vcn = from_phasor_deg(120, 150)

vab = van - vbn
vbc = vbn - vcn
vca = vcn - van

print('vab = ', to_phasor_string(vab))
print('vbc = ', to_phasor_string(vbc))
print('vca = ', to_phasor_string(vca))
```

```python
zs = 0.4 + 0.3j
zload = 24 + 19j
zline = 0.6 + 0.7j
zy = zs + zload + zline

ia = van / zy
ib = vbn / zy
ic = vcn / zy

print('ia = ', to_phasor_string(ia))
print('ib = ', to_phasor_string(ib))
print('ic = ', to_phasor_string(ic))
```

# Practice 3

```python
vp = 120
angle_ab = -20
vab = from_phasor_deg(vp, angle_ab)
z = from_phasor_deg(20, 40)
vbc = from_phasor_deg(vp, angle_ab - 120)
vca = from_phasor_deg(vp, angle_ab - 180)

print('vbc = ', to_phasor_string(vbc))
print('vca = ', to_phasor_string(vca))
```

```python
iab = vab / z
ibc = vbc / z
ica = vca / z

print("iab = ", to_phasor_string(iab))
print("ibc = ", to_phasor_string(ibc))
print('ica = ', to_phasor_string(ica))
```

# Problem 4

```python
z = 18 + 12j
ia = from_phasor_deg(9.609, 35)
va = z * ia
print('va = ', to_phasor_string(va))
```

```python
import numpy as np
ip_mag = abs(ia) / np.sqrt(3)
iab = from_phasor(ip_mag, np.angle(ia) + np.deg2rad(30))
print('iab = ', to_phasor_string(iab))
```

```python
vab = iab * z
print('vab = ', to_phasor_string(vab))
```
